```python
def get_abstract(title):
    time.sleep(2)
    works = Works().search_filter(title=title).get()
    for work in works:
        abstract = work['abstract']
        referenced_works = work['referenced_works']
        id = work['ids']['openalex']
        return abstract,referenced_works, id
    return None,None,None
```

```python
file_path = "E:/document/research/dr_research/code/eswa临时/未获奖.xlsx"
df = pd.read_excel(file_path)
```

```python
from pyalex.api import QueryError
for index, row in df.iterrows():
    print(index)
    if index < 12719:
        continue
    title = row['title']
    try:
        abstract, referenced_works, id= get_abstract(title)
        df.at[index, 'abstract'] = abstract
        df.at[index,"id"] = id
    except QueryError as e:
        continue
    
    # 创建一个来存储 ref
    if referenced_works != None:
        print(index,"ref开始处理")
        ref_data = {'id': [id] * len(referenced_works), 'ref_id': referenced_works}
        df_ref = pd.DataFrame(ref_data)
        folder_path = "E:/document/research/dr_research/code/eswa临时/noaward_ref"
        file_name = f"{folder_path}/{index}_ref.xlsx"
        df_ref.to_excel(file_name, index=False)
        print(index,"ref处理完毕")
```

```python
import time
import pandas as pd
from pyalex import Works
import pyalex

pyalex.config.email = "xiahzaq@163.com"
from pyalex import config

config.max_retries = 0
config.retry_backoff_factor = 0.1
config.retry_http_codes = [429, 500, 503]
```

```python
df = pd.read_excel("E:/document/research/dr_research/code/eswa临时/未获奖.xlsx")
```

```python
from openpyxl.utils.exceptions import IllegalCharacterError

works = Works()
for index, row in df.iterrows():
    # 过滤掉没摘要的
    if pd.isna(row['id']) or pd.isna(row['abstract']):
        continue
        
    print(index, "开始处理")
    if index < 25347:
        continue
        
    try:
        folder_path = "E:/document/research/dr_research/code/eswa临时/noaward_ref"
        file_name = f"{folder_path}/{index}_ref.xlsx"
        df_ref = pd.read_excel(file_name)
    except FileNotFoundError:
        print(f"文件 {file_name} 不存在，跳过处理")
        continue
    
    if df_ref.empty:
        continue
        
    abstracts = []

    for ref_id in df_ref.iloc[:, 1]:
        time.sleep(0.1)
        try:
            work = works[ref_id]
            abstract = work['abstract']
            abstracts.append(abstract)
        except ValueError as e:
            abstracts.append(None)
            continue
        
        except Exception as e:
            abstracts.append(None)
            print(e)
            continue
            
    df_ref['abstract'] = abstracts
    df_ref.to_excel(file_name, index=False)
    print(index, "处理完毕")
```
